# Data Assembly and Evaluation

### Combine the boxscores for the seasons being used.

### Evaluate the fields for completeness (looking at number of NaN values).

### Generate distributions on key statistics to review for problem data.

```python
import pandas as pd
import numpy as np
import glob
import os
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('seaborn-darkgrid')
```

```python
path = r'C:\Users\seanh\college_basketball_ml_project\output\boxscores'                 # use your path
all_files = glob.iglob(os.path.join(path, "*.csv"))     # advisable to use os.path.join as this makes concatenation OS independent

box_hist_df = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True, sort=False)

box_hist_df.head()
```

```python
for col in box_hist_df.columns: 
    print(box_hist_df.dtypes)
```

```python
box_hist_df.home_three_point_field_goals.fillna(0, inplace = True) 
box_hist_df.home_three_point_field_goals = pd.to_numeric(box_hist_df.home_three_point_field_goals,errors='coerce')

box_hist_df.home_three_point_field_goal_attempts.fillna(0, inplace = True) 
box_hist_df.home_three_point_field_goal_attempts = pd.to_numeric(box_hist_df.home_three_point_field_goal_attempts,errors='coerce')

box_hist_df.home_two_point_field_goal_attempts.fillna(0, inplace = True) 
box_hist_df.home_two_point_field_goal_attempts = pd.to_numeric(box_hist_df.home_two_point_field_goal_attempts,errors='coerce')

for col in box_hist_df.columns: 
    print(box_hist_df.dtypes)
```
