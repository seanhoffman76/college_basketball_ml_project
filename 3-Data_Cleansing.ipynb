```python
import numpy as np
import pandas as pd
import gc
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('seaborn-darkgrid')
```

```python
combo_box = 'output/boxscores/2011-2019-combined_boxscores.csv'

box_hist_df = pd.read_csv(combo_box)
box_hist_df.head(20)
```

```python
# Drop the home_ranking and away_ranking fields due to null values (home = 92530 null, away = 95469 null)
# Also need to remove team rankings so as not to skew the model.

box_hist_df = box_hist_df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1','home_ranking','away_ranking'])
box_hist_df.head(20)
```

```python
# Determine the number of teams with a value of '\n\t\t\t' in their name. These are schools outside of NCAA Division I and 
# aren't tracked in the sports-reference.com database. Advanced statistics are unavailable for these teams.

# See example at https://www.sports-reference.com/cbb/boxscores/2010-11-14-air-force.html

teams_list = list(set(list(box_hist_df['winning_name'].values)+list(box_hist_df['losing_name'].values)))
print(teams_list[:50])

teams = []
for t in teams_list:
    if '\n\t\t\t' not in t:
        teams.append(t)

print()
print(len(teams))
print()
print(teams[:50])
```

```python
# Create a function to attach the season information to each game.

def season(df):
    # datetime
    df['date'] = pd.to_datetime(df['date'])
    
    df['month']  = df['date'].copy().dt.month.astype(int)
    df['year']   = df['date'].copy().dt.year.astype(int)
    df['season'] = df['year'].copy()
    df['season'] = np.where(df['month']>9, df['season']+1, df['season'])
    
    df = df.drop(columns=['month','year'])
    
    return df

box_seasons_df = season(box_hist_df)

box_seasons_df.head(20)
```

```python
# Create a function to assign the day of the season in which each game was played.

def day_of_season_func(df):
    
    df['season_start'] = df['season'].copy() - 1
    df['season_start'] = df['season_start'].copy().astype(str) + '-10-31'
    
    df['day_of_season'] = pd.to_datetime(df['date']) - pd.to_datetime(df['season_start'])
    df['day_of_season'] = df['day_of_season'].dt.days.astype(int)
    
    return df

box_dos_df = day_of_season_func(box_seasons_df)

print(box_dos_df.day_of_season.min())
print(box_dos_df.day_of_season.max())

box_dos_df.head(20)
```

```python
# Create a function to generate game ID values

def add_id_func(df):
    
    df['wname_copy'] = df['winning_name'].str.replace('\n\t\t\t','').str.replace(' ','').str.lower()
    df['lname_copy'] = df['losing_name'].str.replace('\n\t\t\t','').str.replace(' ','').str.lower()
    
    df['wname_copy2'] = df['wname_copy'].str[:3]
    df['wname_copy3'] = df['wname_copy'].str[-3:]
    
    df['lname_copy2'] = df['lname_copy'].str[:3]
    df['lname_copy3'] = df['lname_copy'].str[-3:]
    
    df['game_id'] = df['date'].dt.strftime('%m%d%y') + df['wname_copy2'] + df['wname_copy3'] + df['lname_copy2']  + df['lname_copy3']
    
    df = df.drop_duplicates(subset=['game_id'])
    
    df = df.drop(columns=['wname_copy2','lname_copy2','wname_copy3','lname_copy3'])
    
    df['team1_id'] = df['season'].astype(str)+df['wname_copy']
    df['team2_id'] = df['season'].astype(str)+df['lname_copy']
    
    df['alt1_id'] = df['wname_copy']
    df['alt2_id'] = df['lname_copy']
    
    df2 = df.copy()
    
    df  =  df.drop(columns=['team1_id'])
    df2 = df2.drop(columns=['team2_id'])
    df  =  df.drop(columns=['alt1_id'])
    df2 = df2.drop(columns=['alt2_id'])
    
    df  =  df.rename(columns={'team2_id':'team_id','alt2_id':'alt_id'})
    df2 = df2.rename(columns={'team1_id':'team_id','alt1_id':'alt_id'})
    
    df = pd.concat([df,df2],axis=0)
    
    df['home_id'] = np.where(df['winner']=='Home', df['season'].astype(str)+df['wname_copy'], df['season'].astype(str)+df['lname_copy'])
    df['away_id'] = np.where(df['winner']=='Away', df['season'].astype(str)+df['wname_copy'], df['season'].astype(str)+df['lname_copy'])
    
    df['win_id']  = df['season'].astype(str)+df['wname_copy']
    df['lose_id'] = df['season'].astype(str)+df['lname_copy']
    
    df = df.drop(columns=['wname_copy','lname_copy'])
    
    print(len(df))
    
    return df

box_dosids_df = add_id_func(box_dos_df)

box_dosids_df[['home_id','away_id','alt_id']].head(20)
```

```python
# Create team specific ratings leading up to game.
# Start with process to show team and opponent rather than home and away.

home_games = box_dosids_df.loc[box_dosids_df['home_id']==box_dosids_df['team_id']]
away_games = box_dosids_df.loc[box_dosids_df['away_id']==box_dosids_df['team_id']]

del box_dosids_df
gc.collect()
```

```python
# Convert column names to reflect team and opponent rather than home and away.
# Process a loop first for the home games.

old_cols = list(home_games)
home_cols = []

for oc in old_cols:
    if 'away_' in oc:
        hc = oc.replace('away_','opp_')
    elif 'home_' in oc:
        hc = oc.replace('home_','')
    else:
        hc = oc
    home_cols.append(hc)
    
# Process an opposite loop for away games
away_cols = []

for oc in old_cols:
    if 'home_' in oc:
        ac = oc.replace('home_','opp_')
    elif 'away_' in oc:
        ac = oc.replace('away_','')
    else:
        ac = oc
    away_cols.append(ac)
    
home_games.columns=home_cols
away_games.columns=away_cols

col_order = list(away_games)
home_games = home_games[col_order]

box_convert_df = pd.concat([home_games,away_games], axis=0)

# Remove a duplicate team_id column
box_convert_df = box_convert_df.drop(columns=['team_id'])

del home_games
del away_games
gc.collect()
```

```python
# Create a function that will calculate the exponential weighted mean by team for features to create 5 year averages
# Span will equal 5 seasons, weighted averages are calculated recursively 

def team_exwtmean(team, span=5, alpha=0.85):
    feature_exwtmean = team.rolling(window=span, min_periods=1).mean()[:span]
    rest = team[span:]
    return pd.concat([feature_exwtmean, rest]).ewm(alpha=alpha, adjust=False).mean()
```

```python
# Add season by season statistics for 'two_point_field_goal_percentage','three_point_field_goal_percentage', 
# 'free_throw_percentage', 'opp_two_point_field_goal_percentage','opp_three_point_field_goal_percentage', 
# 'opp_free_throw_percentage'

box_calc = box_convert_df.groupby(['season','alt_id'])['offensive_rating','pace','opp_two_point_field_goal_percentage','free_throw_percentage','three_point_field_goal_percentage','steal_percentage'].mean().reset_index()

box_calc[['Off_Rtg','Pace','Opp_FGM%','FTM%','3FGM%','Stl%']] = box_calc.groupby(['alt_id'])['offensive_rating','pace','opp_two_point_field_goal_percentage','free_throw_percentage','three_point_field_goal_percentage','steal_percentage'].shift()

box_calc = box_calc.dropna()

box_calc[['5yr_Off_Rtg','5yr_Pace','5yr_Opp_FGM%','5yr_FTM%','5yr_3FGM%','5yr_Stl%']] = box_calc.groupby(['alt_id'])['Off_Rtg','Pace','Opp_FGM%','FTM%','3FGM%','Stl%'].transform(team_exwtmean)

box_calc['id'] = box_calc['season'].astype(str)+box_calc['alt_id']

box_calc = box_calc[['id','5yr_Off_Rtg','5yr_Pace','5yr_Opp_FGM%','5yr_FTM%','5yr_3FGM%','5yr_Stl%']]

box_calc.head(20)
```

```python
# Add season by season statistics for 'free_throws','two_point_field_goals','three_point_field_goals',
# 'offensive_rebounds','defensive_rebounds','opp_offensive_rebounds','opp_defensive_rebounds',
# 'turnovers','wins','losses'

box_calc_more = box_convert_df.groupby(['season','alt_id'])['free_throws','two_point_field_goals','three_point_field_goals','offensive_rebounds','defensive_rebounds','opp_offensive_rebounds','opp_defensive_rebounds','turnovers','wins','losses','pace'].sum().reset_index()

box_calc_more[['fts','2pt','3pt','oreb','dreb','oor','odr','tos','w','l','pace']] = box_calc_more.groupby(['alt_id'])['free_throws','two_point_field_goals','three_point_field_goals','offensive_rebounds','defensive_rebounds','opp_offensive_rebounds','opp_defensive_rebounds','turnovers','wins','losses','pace'].shift()

box_calc_more = box_calc_more.dropna()

# Calculate defensive rebounds available
box_calc_more['dra'] = box_calc_more['dreb'] + box_calc_more['oor']

# Calculate offensive rebounds available
box_calc_more['ora'] = box_calc_more['oreb'] + box_calc_more['odr']

# Calculate offensive/defensive rebound percentages

box_calc_more['DRebPct'] = box_calc_more['dreb'] / box_calc_more['dra']
box_calc_more['ORebPct'] = box_calc_more['oreb'] / box_calc_more['ora']
box_calc_more['_points'] = box_calc_more['fts'] + 2 * box_calc_more['2pt'] + 3 * box_calc_more['3pt']
box_calc_more['FTPct']   = box_calc_more['fts'] / box_calc_more['_points']
box_calc_more['2ptPct']  = box_calc_more['2pt'] / box_calc_more['_points']
box_calc_more['3ptPct']  = box_calc_more['3pt'] / box_calc_more['_points']
box_calc_more['TO%']     = box_calc_more['tos'] / box_calc_more['pace']
box_calc_more['gp']      = box_calc_more['w'] + box_calc_more['l']
box_calc_more['W%']      = box_calc_more['w'] / box_calc_more['gp']

box_calc_more[['5y_DReb','5y_OReb','5y_FTPct','5y_2PT','5y_3PT','5y_TO','5y_Win']] = box_calc_more.groupby(['alt_id'])['DRebPct','ORebPct','FTPct','2ptPct','3ptPct','TO%','W%'].transform(team_exwtmean)

box_calc_more['id'] = box_calc_more['season'].astype(str) + box_calc_more['alt_id']

box_calc_more = box_calc_more[['id','5y_OReb','5y_FTPct','5y_3PT','5y_TO','5y_Win']]

box_calc_more.head(20)
```

```python
# Create a function that will calculate the exponential weighted mean by team for features to create season-to-date averages
# Span will cover most recent 50 games in the season, weighted averages are calculated recursively 

def team_exwtmean_game(team, span=50, alpha=0.84):
    feature_exwtmean = team.rolling(window=span, min_periods=1).mean()[:span]
    rest = team[span:]
    return pd.concat([feature_exwtmean, rest]).ewm(alpha=alpha, adjust=False).mean()
```

```python
exwtmean = box_convert_df.groupby(['id'])['ORtg','DRtg','Pace','Ast%','Blk%','PF','Stl%','FGM%','3FGM%','FTM%','Opp_Ast%','Opp_Blk%','Opp_PF','Opp_Stl%','Opp_FGM%','Opp_3FGM%','Opp_FTM%'].apply(team_exwtmean_game).reset_index()

if 'level_1' in list(exwtmean):
    exwtmean = exwtmean.drop(columns=['level_1'])

exwtmean.columns=['id','wORtg','wDRtg','wPace','wAst%','wBlk%','wPF','wStl%','wFGM%','w3FGM%','wFTM%','wOpp_Ast%','wOpp_Blk%','wOpp_PF','wOpp_Stl%','wOpp_FGM%','wOpp_3FGM%','wOpp_FTM%']

boxcalc_wt = pd.concat([expdmean, exwtmean.drop(columns=['id'])], axis=1)

boxcalc_wt.columns=['id','Sea_ORtg','Sea_DRtg','Sea_Pace','Sea_Ast%','Sea_Blk%','Sea_PF','Sea_Stl%','Sea_FGM%','Sea_3FGM%','Sea_FTM%','Sea_Opp_Ast%','Sea_Opp_Blk%','Sea_Opp_PF','Sea_Opp_Stl%','Sea_Opp_FGM%','Sea_Opp_3FGM%','Sea_Opp_FTM%','wSea_ORtg','wSea_DRtg','wSea_Pace','wSea_Ast%','wSea_Blk%','wSea_PF','wSea_Stl%','wSea_FGM%','wSea_3FGM%','wSea_FTM%','wSea_Opp_Ast%','wSea_Opp_Blk%','wSea_Opp_PF','wSea_Opp_Stl%','wSea_Opp_FGM%','wSea_Opp_3FGM%','wSea_Opp_FTM%']

boxcalc_wt = boxcalc_wt[['id','Sea_DRtg','Sea_Opp_3FGM%','Sea_Opp_PF','Sea_Pace','wSea_ORtg','wSea_Stl%','wSea_Ast%','wSea_Opp_Ast%','wSea_Blk%','wSea_FGM%','Sea_FTM%','Sea_PF']]

boxcalc_wt.head(20)
```

```python
box_convert_df[['fts','2pt','3pt','oreb','dreb','oor','odr','tos','w','l','Pace']] = box_convert_df.groupby(['id'])['free_throws','two_point_field_goals','three_point_field_goals','offensive_rebounds','defensive_rebounds','opp_offensive_rebounds','opp_defensive_rebounds','turnovers','wins','losses','pace'].shift()

expdsum = box_convert_df.groupby(['id'])['fts','2pt','3pt','oreb','dreb','oor','odr','tos','w','l','Pace'].expanding().sum().reset_index()

if 'level_1' in list(expdsum):
    expdsum = expdsum.drop(columns=['level_1'])
    
    
# Calculate defensive rebounds available
expdsum['dra'] = expdsum['dreb'] + expdsum['oor']

# Calculate offensive rebounds available
expdsum['ora'] = expdsum['oreb'] + expdsum['odr']

# Calculate offensive/defensive rebound percentages

expdsum['DRebPct'] = expdsum['dreb'] / expdsum['dra']
expdsum['ORebPct'] = expdsum['oreb'] / expdsum['ora']
expdsum['_points'] = expdsum['fts'] + 2 * expdsum['2pt'] + 3 * expdsum['3pt']
expdsum['FTPct']   = expdsum['fts'] / expdsum['_points']
expdsum['2ptPct']  = expdsum['2pt'] / expdsum['_points']
expdsum['3ptPct']  = expdsum['3pt'] / expdsum['_points']
expdsum['TO%']     = expdsum['tos'] / expdsum['Pace']
expdsum['gp']      = expdsum['w'] + expdsum['l']
expdsum['W%']      = expdsum['w'] / expdsum['gp']

expdsum.head(20)
```

```python
# Weighted Sum

def team_wtsum(team, span=50, alpha=0.84):
    feature_ewm = team.rolling(window=span, min_periods=1).sum()[:span]
    rest = team[span:]
    return pd.concat([feature_ewm, rest]).ewm(alpha=alpha, adjust=False).mean()


ews = box_convert_df.groupby(['id'])['fts','2pt','3pt','oreb','dreb','oor','odr','tos','w','l','Pace'].apply(team_wtsum).reset_index()

if 'level_1' in list(ews):
    ews = ews.drop(columns=['level_1'])
    
    
# Calculate defensive rebounds available
ews['dra'] = ews['dreb'] + ews['oor']

# Calculate offensive rebounds available
ews['ora'] = ews['oreb'] + ews['odr']

# Calculate offensive/defensive rebound percentages

ews['DRebPct'] = ews['dreb']/ews['dra']
ews['ORebPct'] = ews['oreb']/ews['ora']
ews['_points'] = ews['fts'] + 2*ews['2pt'] + 3*ews['3pt']
ews['FTPct'] = ews['fts']/ews['_points']
ews['2ptPct'] = ews['2pt']/ews['_points']
ews['3ptPct'] = ews['3pt']/ews['_points']
ews['TO%'] = ews['tos']/ews['Pace']
ews['gp'] = ews['w'] + ews['l']
ews['W%'] = ews['w']/ews['gp']

expdsum = expdsum[['DRebPct','ORebPct','FTPct','2ptPct','3ptPct','TO%','W%']]
ews = ews[['DRebPct','ORebPct','FTPct','2ptPct','3ptPct','TO%','W%']]

expdsum.columns= ['Sea_DReb','Sea_OReb','Sea_FT','Sea_2pt','Sea_3pt','Sea_TO','Sea_W']
ews.columns= ['wSea_DReb','wSea_OReb','wSea_FT','wSea_2pt','wSea_3pt','wSea_TO','wSea_W']

to_add = pd.concat([expdsum, ews], axis=1)

to_add = to_add[['Sea_W','Sea_3pt','Sea_FT','Sea_TO']]

boxcalc_wt = pd.concat([boxcalc_wt, to_add], axis=1)

boxcalc_wt.head(20)
```
